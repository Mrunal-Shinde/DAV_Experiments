# **Matplotlib:**
Matplotlib is a popular plotting library in Python used for creating high-quality visualizations and graphs. It offers various tools to generate diverse plots, facilitating data analysis, exploration, and presentation. Matplotlib is flexible, supporting multiple plot types and customization options, making it valuable for scientific research, data analysis, and visual communication.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 1, 0.05)
y = x**2

fig, ax = plt.subplots(figsize=(7,4))
plt.scatter(x, y)
ax.grid(True, linestyle='-.', linewidth=1, color='red', alpha=0.9)
ax.set_title('Customizing Grids')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category A', 'Category B', 'Category C']
values1 = [15, 24, 30]
values2 = [20, 18, 25]

bar_width = 0.35
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width

plt.bar(bar_positions1, values1, width=bar_width, label='Group 1', color='skyblue')
plt.bar(bar_positions2, values2, width=bar_width, label='Group 2', color='orange')

plt.xlabel('Categories')
plt.ylabel('Values')

plt.title('Grouped Bar Graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
explode = (0.1, 0, 0.2, 0.3, 0)
colors = ('orange', 'cyan', 'brown', 'grey', 'indigo')
wp = {'linewidth': 1, 'edgecolor': 'green'}

def func(pct, allvalues):
    absolute = int(pct / 100. * np.sum(allvalues))
    return "{:.1f}%\n({:d} students)".format(pct, absolute)

fig, ax = plt.subplots(figsize=(6, 4))
wedges, texts, autotexts = ax.pie(students,
                                   autopct=lambda pct: func(pct, students),
                                   explode=explode,
                                   labels=langs,
                                   shadow=True,
                                   colors=colors,
                                   startangle=140,
                                   wedgeprops=wp)

ax.legend(wedges, langs,
          title='Languages',
          loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=6, weight='bold', color='magenta')

ax.set_title('pie chart')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generating data for a spherical surface
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, np.pi, 100)
theta, phi = np.meshgrid(theta, phi)
r = 1
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

# Creating a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Plotting the spherical wireframe
ax.plot_wireframe(x, y, z, color='blue')

# Adding labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Basic 3D Wireframe Plot')

# Displaying the plot
plt.show()
```

```python
import pandas as pd
# https://www.datacamp.com/tutorial/matplotlib-tutorial-python
# dataset_link = https://www.wsj.com/market-data/quotes/index/DJIA/historical-prices
djia_data = pd.read_csv('HistoricalPrices.csv')
djia_data.head()
```

```python
djia_data = djia_data.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
djia_data['Date'] = pd.to_datetime(djia_data['Date'])
djia_data = djia_data.sort_values(by ='Date')
```

```python
plt.plot(djia_data['Date'], djia_data['Open'], label = 'Open', color = 'black')
plt.plot(djia_data['Date'], djia_data['Close'], label = 'Close', color = 'red')
plt.xlabel("Date")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
from calendar import month_name

# Order by months by chronological order
djia_data['Month'] = pd.Categorical(djia_data['Date'].dt.month_name(), month_name[1:])

# Group metrics by monthly averages
djia_monthly_mean = djia_data \
    .groupby('Month') \
    .mean(numeric_only=True) \
    .reset_index()

djia_monthly_mean.head(6)
```

```python
plt.bar(djia_monthly_mean['Month'], height = djia_monthly_mean['Close'])
plt.xticks(rotation=45)
plt.show()
```

```python
z = np.polyfit(djia_data['Open'], djia_data['Close'], 1)
p = np.poly1d(z)

plt.xlabel("Open Prices")
plt.ylabel("Closed Prices")
plt.title("DJIA 2022 Scatterplot Open vs. Close")
plt.plot(djia_data['Open'], p(djia_data['Open']))
plt.scatter(djia_data['Open'], djia_data['Close'], c=djia_data['Close'], cmap = plt.cm.plasma)
plt.show()
plt.savefig('DJIA 2022 Scatterplot Open vs. Close.png')
```
