## **Python**

**Explore Top-5 Data Visualization Libraries in Python (w.r.t Features & Applications)**



1.   **Matplotlib:**
Matplotlib is a widely-used Python library for creating static, interactive, and publication-quality visualizations. It offers a vast array of plotting functions, allowing users to generate a variety of plots including line plots, scatter plots, histograms, and more. With its flexibility and customization options, Matplotlib is favored by scientists, engineers, and data analysts for data exploration and presentation.

2.   **Seaborn:**
Seaborn is a Python data visualization library that builds on top of Matplotlib and provides a high-level interface for creating attractive statistical graphics. It simplifies the process of generating complex visualizations such as distribution plots, categorical plots, and relational plots. Seaborn's ease of use and aesthetic appeal make it a valuable tool for data analysis and exploration.

3.   **Plotnine (ggplot):**
Plotnine is a Python implementation of the ggplot2 library in R, offering a grammar of graphics approach to visualization. It enables users to create complex plots by mapping data to visual elements in a concise and readable manner. Plotnine's adherence to the principles of the Grammar of Graphics facilitates reproducibility and clarity in code, making it a popular choice for creating publication-quality visualizations in Python.

4.   **Bokeh:**
Bokeh is a Python library for creating interactive visualizations that are designed to be displayed in modern web browsers. It provides a flexible and powerful framework for building interactive plots, dashboards, and data applications. Bokeh's interactive features, including widgets and streaming data capabilities, make it well-suited for developing dynamic and engaging web-based data visualizations.

5.   **Pygal:**
Pygal is a lightweight Python library for creating SVG-based charts and graphs. It offers a simple and intuitive API for generating various types of charts including line charts, bar charts, and pie charts. Pygal's focus on simplicity and scalability makes it a convenient choice for generating static and interactive visualizations that can be easily embedded into web applications or documents.

**Perform simple experiments on the 2 identified libraries for Python**

```r
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("/content/diabetes.csv")
df.columns
df.describe()
```

```r
plt.figure(figsize=(12,12))
plt.subplot(3,3,1)
sns.barplot(x='Outcome',y='Glucose',data=df,hue="Outcome")
plt.subplot(3,3,2)
sns.barplot(x='Outcome',y='BloodPressure',data=df,hue="Outcome")
plt.subplot(3,3,3)
sns.barplot(x='Outcome',y='SkinThickness',data=df,hue="Outcome")
plt.subplot(3,3,4)
sns.barplot(x='Outcome',y='BMI',data=df,hue="Outcome")
plt.subplot(3,3,5)
sns.barplot(x='Outcome',y='DiabetesPedigreeFunction',data=df,hue="Outcome")
plt.subplot(3,3,6)
sns.barplot(x='Outcome',y='Age',data=df,hue="Outcome")
```

```r
sns.catplot(x='Pregnancies',y='Insulin',data=df,hue='Outcome')
```

```r
sns.boxplot(x="Pregnancies",y="Age",data=df,hue="Outcome")
```

```r
plt.hist(df['Age'], bins=20, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age in Diabetes Dataset')
plt.show()
```

```r
plt.scatter(df['BMI'], df['Glucose'], color='green', alpha=0.5)
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.title('Scatter Plot: BMI vs. Glucose')
plt.show()
```

## **R**

**Explore Top-5 Data Visualization Libraries in R (w.r.t Features & Applications)**



1.   **RGL:**
RGL, or "R Graphics Library," is a powerful package for creating 3D visualizations in R. It provides functions to create interactive 3D plots and animations directly in R. With RGL, users can create complex 3D scenes, manipulate objects, apply lighting effects, and interactively explore data in three dimensions.

2.   **Dygraphs:**
Dygraphs is an R package for creating interactive time series plots and charts. It is built on the principles of the JavaScript library dygraphs.js and enables users to visualize time series data with interactive features such as zooming, panning, and highlighting specific data points. Dygraphs supports various types of time series visualizations including line charts, bar charts, and area charts.

3.   **Leaflet:**
Leaflet is an R package for creating interactive maps and web-based geographic visualizations. It integrates the Leaflet JavaScript library with R, allowing users to create dynamic and interactive maps directly in R scripts and Shiny applications. Leaflet provides functions to add layers, markers, pop-ups, and controls to maps, making it easy to customize the appearance and behavior of maps.

4.   **ggplot2:**
ggplot2 is a popular R package for creating elegant and versatile graphics based on the grammar of graphics. It provides a consistent and flexible syntax for building complex visualizations by mapping data to aesthetic attributes such as color, shape, size, and position. ggplot2 follows the layered approach to constructing plots, allowing users to add multiple layers of data and geometric objects to create sophisticated visualizations.

5.  **Lattice:**
Lattice is an R package for creating conditioned plots and trellis displays. It provides a powerful framework for visualizing multivariate data by creating panels or small multiples based on conditioning variables. Lattice plots are created using high-level functions that take formula syntax, making it easy to specify relationships between variables and create customized displays for different subsets of data.

```r
library("ggplot2")
data=read.csv("diabetes.csv")
```

```r
ggplot(data, aes(x=BMI, y=Glucose)) +
  geom_point(color="green", alpha=0.5) +
  labs(title="Scatter Plot: BMI vs. Glucose", x="BMI", y="Glucose")
```

```r
ggplot(data, aes(x=Age)) +
  geom_histogram(fill="blue", color="black", bins = 20) +
  labs(title="Distribution of Age in Diabetes Dataset", x="Age", y="Frequency")
```

```r
library("lattice")
data=read.csv("diabetes.csv")
```

```r
histogram(~ Age, data = data,
          xlab = "Age",
          ylab = "Frequency",
          main = "Distribution of Age in Diabetes Dataset",
          col = "orange")
```

```r
xyplot(BMI ~ Glucose, data = data,
       xlab = "Glucose",
       ylab = "BMI",
       main = "Scatter Plot: BMI vs. Glucose",
       col = "violet")
```

```r
age <- data$Age

stripchart(age,
           method = "jitter",
           main = "Stripchart of Age in Diabetes Dataset",
           xlab = "Age",
           ylab = "Frequency",
           pch = 19,
           col = "blue"
)
```
